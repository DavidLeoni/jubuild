# Jupman

**Jupman is a template manager for course websites made with** [Jupyter](http://jupyter.org) **notebooks and** [NBSphinx](http://nbsphinx.readthedocs.io/) **doc generator**.

**Features**: 

- automated generation of exercises from solution templates (both .ipynb and .py)
- automated generation of chapter zips
- supports sharing code among chapters
- produces a website and a decent PDF 
- Python Tutor integration (can work offline, doesn't need to install dependencies)
- includes a exam management system (script and grades spreadsheet)
- configuration clearly separated from code
- made for Python 3
- comes with [documentation](usage.ipynb)
- Open source code [on Github](https://github.com/DavidLeoni/jupman)
- Apache License v2.0

**Currently lacking:**

- decent EPUB support
- more testing, especially for exam management

**Used by:**

- [Scientific Programming Lab at University of Trento, Data Science Master](http://datasciprolab.readthedocs.io/) (English)
- [SoftPython book](http://softpython.readthedocs.io/) (Italian)
